Spam Email Classifier

This notebook demonstrates how to build a simple NLP model to classify emails as 'Spam' or 'Ham' (not spam).

```python
import pandas as pd
import numpy
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

1. Load Data

```python
df=pd.read_csv('emails.csv')
```

```python
df
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
x=df['text']
y=df['spam']
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

2. Model Pipeline

We will use TfidfVectorizer to convert text into numerical features & MultinomialNB (Naive Bayes) and Support Vector Machines (SVM)  for classification..

```python
vectorizer = TfidfVectorizer(stop_words='english', max_df=0.7)
```

```python
X_train_tfidf = vectorizer.fit_transform(X_train)
X_test_tfidf = vectorizer.transform(X_test)
```

```python
print("Training Naive Bayes Classifier")
nb_model = MultinomialNB()
nb_model.fit(X_train_tfidf, y_train)
```

```python
nb_predictions = nb_model.predict(X_test_tfidf)
print(f"Naive Bayes Accuracy: {accuracy_score(y_test, nb_predictions):.4f}")
print("Classification Report:")
print(classification_report(y_test, nb_predictions))
```

```python
print("Training SVM Classifier")
svm_model = SVC(kernel='rbf',random_state=42)
svm_model.fit(X_train_tfidf, y_train)
```

```python
svm_predictions = svm_model.predict(X_test_tfidf)
print(f"SVM Accuracy: {accuracy_score(y_test, svm_predictions):.4f}")
print("Classification Report:")
print(classification_report(y_test, svm_predictions))
```

3. Testing with Custom User Inputs

```python

custom_emails = [
    'Hey, are we still on for the movie?',
    'Click here to claim your $1000 prize now!'
]
custom_emails_tfidf = vectorizer.transform(custom_emails)

nb_custom_predictions = nb_model.predict(custom_emails_tfidf)
svm_custom_predictions = svm_model.predict(custom_emails_tfidf)

print("\nCustom Input Test Results")
for i, email in enumerate(custom_emails):
    print(f"\n[Email #{i+1}]: {email}")
    
    # Map 1 to 'SPAM' and 0 to 'HAM'
    nb_result = "SPAM" if nb_custom_predictions[i] == 1 else "HAM (Normal)"
    svm_result = "SPAM" if svm_custom_predictions[i] == 1 else "HAM (Normal)"
    
    print(f" └─ Naive Bayes Predicts: {nb_result}")
    print(f" └─ SVM Predicts:         {svm_result}")
```
